```python
import os
import pdfplumber

data_directory = "../notebook/data/pdf_files"

# 지정된 디렉토리의 모든 파일 리스트 가져오기
file_list = os.listdir(data_directory)

# PDF 파일만 필터링
pdf_files = [file for file in file_list if file.endswith('.pdf')]

# 첫 번째 PDF 파일 하나 선택
if pdf_files:
    first_pdf_file = pdf_files[1]
    print(first_pdf_file)
    pdf_file_path = os.path.join(data_directory, first_pdf_file)

    # pdfplumber로 PDF 파일 열기
    with pdfplumber.open(pdf_file_path) as pdf:
        # 각 페이지를 순회하며 텍스트와 표 추출
        for page_number, page in enumerate(pdf.pages, start=1):
            print(f"--- Page {page_number} ---")

            # 페이지 텍스트 추출
            text = page.extract_text()
            if text:
                print("텍스트:")
                print(text)  # 페이지의 텍스트를 출력
            else:
                print("텍스트가 없습니다.")

else:
    print("PDF 파일이 없습니다.")
```
